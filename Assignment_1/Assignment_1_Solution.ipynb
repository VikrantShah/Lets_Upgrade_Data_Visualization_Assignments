# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Question 1
Plot a line plot between a and b:<br><br>
a=np.arange(40,50)<br>
b=np.arange(50,60)


```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.plot(a, b, "bo-", linewidth=3, markersize=6)
plt.title("Question 1 Solution")
plt.xlabel("a")
plt.ylabel("b")
plt.savefig("Question_1_Solution.jpg")
plt.show()
```

# Question 2
Plot a line plot showing the sales trend in company 1 and 2:<br><br>
days = [1,2,3,4,5,6,7] #days of d week<br><br>
sales_1 = [160,150,140,145,175,165,180] #sales of company1<br>
sales_2 = [70,90,160,150,140,145,175]  #sales of company2


```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2
```

```python
plt.plot(days, sales_1, "r--", label="Sales of Company 1")
plt.plot(days, sales_2, "b-.", label="Sales of Company 2")
plt.title("The Sales Trend in Company 1 & Company 2")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.savefig("Question_2_Solution.jpg")
plt.show()
```

# Question 3
Create a 3 by 3 subplots:<br>
multiple plots <br><br>
x = [1,2,3,4]<br><br>
y1 = [4,3,2,1]<br>
y2 = [10,20,30,40]<br>
y3 = [40,30,20,10]<br>
y4 = [1,2,1,2]<br>
y5 = [40,70,90,70]<br>


```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.figure(figsize = (12, 12))

plt.subplot(3, 3, 1)
plt.plot(x, y1, "ro")
plt.xlabel("x")
plt.ylabel("y1")
plt.title("(x, y1)")

plt.subplot(3, 3, 2)
plt.plot(x, y2, "g-")
plt.xlabel("x")
plt.ylabel("y2")
plt.title("(x, y2)")

plt.subplot(3, 3, 3)
plt.plot(x, y3, "b--")
plt.xlabel("x")
plt.ylabel("y3")
plt.title("(x, y3)")

plt.subplot(3, 3, 4)
plt.plot(x, y4, "c-.")
plt.xlabel("x")
plt.ylabel("y4")
plt.title("(x, y4)")

plt.subplot(3, 3, 5)
plt.plot(x, y5, "k:")
plt.xlabel("x")
plt.ylabel("y5")
plt.title("(x, y5)")

plt.savefig("Question_3_Solution.jpg")
plt.show()
```
